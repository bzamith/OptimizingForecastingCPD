```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("datasets/embrapa/pantanal_climate.csv")
df.head()
```

```python
g = sns.lineplot(data=df, x='Date', y='P')
_ = g.set_xticklabels(g.get_xticklabels(), rotation=30)
plt.gcf().set_size_inches(40,20)
```

```python
g = sns.lineplot(data=df[df['Date'] > '2017-01-01'], x='Date', y='P')
_ = g.set_xticklabels(g.get_xticklabels(), rotation=30)
plt.gcf().set_size_inches(40,20)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
per = pd.DatetimeIndex(df.Date).to_period("M")
df_2 = df.groupby(per).sum()
df_2['Date'] = df_2.index
df_2.head()
```

```python
g = sns.lineplot(data=df_2, x=df_2['Date'].astype(str), y='P')
_ = g.set_xticklabels(g.get_xticklabels(), rotation=30)
plt.gcf().set_size_inches(40,20)
```
